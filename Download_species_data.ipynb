**Download first specie from gbif and save to csv - Amelanchier alnifolia**

```python
import pandas as pd

species_name = 'Amelanchier alnifolia'
spec_info = species.name_backbone(name=species_name)
spec_key = spec_info['speciesKey']

all_results = []


offset = 0
limit = 300

while True:
    spec_occs = occ.search(taxonKey=spec_key, limit=limit, offset=offset)
    results = spec_occs['results']


    all_results.extend(results)


    offset += limit


    if len(results) < limit:
        break


df_Amelanchier = pd.DataFrame(all_results)

df_Amelanchier.shape
```

```python
df_Amelanchier.columns
```

```python

selected_columns = ['scientificName', 'countryCode', 'individualCount', 'decimalLatitude',
                    'decimalLongitude', 'coordinateUncertaintyInMeters', 'year', 'basisOfRecord', 'institutionCode']
selected_data = df_Amelanchier[selected_columns]


new_df = pd.DataFrame(selected_data)

new_df.head(3)
```

```python

df_species1 = new_df.dropna(subset=['decimalLatitude', 'decimalLongitude'])
df_species1.shape
```

```python
df_species1.head(7)
```

```python
df_species1.to_csv('df_species1.csv', index=True)


from google.colab import files
files.download('df_species1.csv')
```
